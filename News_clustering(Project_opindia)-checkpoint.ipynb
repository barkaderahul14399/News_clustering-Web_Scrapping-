```python
#Import Required Modules
```

```python
import pandas as pd
import bs4
import requests
```

```python
#Selecting Website
```

```python
url="https://www.opindia.com/"
```

```python
#getting Data From web
```

```python
data=requests.get(url)
```

```python
data.content
```

```python
# Converting Data into Proper HTML Format
```

```python
soup=bs4.BeautifulSoup(data.content)
```

```python
soup
```

```python
# fetching All urls from Data
```

```python
urls=[]
for i in soup.find_all("h3",class_="entry-title td-module-title"):
    urls.append(i.a["href"])
```

```python
urls
```

```python
# Fetching Headlines From Websites
```

```python
headlines=[]
for i in urls:
    data=requests.get(i)
    soup=bs4.BeautifulSoup(data.content)
    for j in soup.find_all("h1",class_="tdb-title-text"):
        headlines.append(j.text)
```

```python
headlines
```

```python
Articles=[]
for i in urls:
    data=requests.get(i)
    soup=bs4.BeautifulSoup(data.content,"html.parser")
    for j in soup.find_all("h3",class_=["entry-title td-module-title"]):
         Articles.append(j.find('a')['title'])     
```

```python
Articles
```

```python
# Converting Articles And News In DataFrame
```

```python
DF=pd.DataFrame([headlines,Articles]).T
```

# Tokenizing And Stemming Articles

```python
P_art=[]
for i in Articles:
    q=i.upper()
    import re
    q=re.sub("[^A-Za-z0-9 ]"," ",q)
    tk_q=q.split(" ")
    sent= ""
    for j in tk_q:
        from nltk.stem import PorterStemmer
        ps=PorterStemmer()
        sent=sent + " " + ps.stem(j).upper()
    P_art.append(sent)
    
```

```python
P_art
```

```python
tk_q
```

# Making  TFIDF Vector

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer()
A=tf.fit_transform(P_art).toarray()
```

```python
A
```

```python
from sklearn.cluster import KMeans
KMN=KMeans(n_clusters=4)
model=KMN.fit(A)
```

```python
model.labels_
```

```python
DF["cluster"]=model.labels_
```

```python
DF.columns=["Headlines","Articles","Clusters"]
```

```python
DF
```

```python
R={ 0:"Regional",
   1:"Sports",
   2:"Politics",
   3:"Social",
   4:"Religious"
   
}
```

```python
desc=[]
for i in DF.Clusters:
    desc.append(R[i])
```

```python
DF["Description"]=desc
```
